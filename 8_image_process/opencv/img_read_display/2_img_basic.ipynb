## 1.2 图像显示：灰度图

```python
img = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)# IMREAD_GRAYSCALE, IMREAD_COLOR
```

```python
%matplotlib inline

from matplotlib import pyplot as plt

plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

## 1.3 图像显示：彩色图

```python
# OpenCV follows BGR order, while matplotlib likely follows RGB order
# more details here: http://stackoverflow.com/questions/15072736/extracting-a-region-from-an-image-using-slicing-in-python-opencv/15074748#15074748
img = cv2.imread('lena.jpg', cv2.IMREAD_COLOR)# IMREAD_GRAYSCALE, IMREAD_COLOR

img2 = img[:,:,::-1]
```

```python
%matplotlib inline

from matplotlib import pyplot as plt

plt.imshow(img2)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

# 2. 查看图像二维矩阵

## 2.1 灰度图

```python
img = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)# IMREAD_GRAYSCALE, IMREAD_COLOR
```

```python
%matplotlib inline

from matplotlib import pyplot as plt

plt.subplot(1, 2, 1) 
plt.title('original')
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()


plt.subplot(1, 2, 2) 
plt.title('scaled')
plt.imshow(img[50:150, 100:300], cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
