# resize x 2

```python
%matplotlib inline
import cv2 
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
# method-1
img = cv2.imread('img_src_rotated.png') 
res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)

plt.figure(figsize=(10,5))
sub = plt.subplot(1,2,1);sub.imshow(img, cmap=cm.binary)
sub = plt.subplot(1,2,2);sub.imshow(res, cmap=cm.binary)
plt.show()
```

```python
# method-2
img = cv2.imread('img_src_rotated.png') 
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)

plt.figure(figsize=(10,5))
sub = plt.subplot(1,2,1);sub.imshow(img, cmap=cm.binary)
sub = plt.subplot(1,2,2);sub.imshow(res, cmap=cm.binary)
plt.show()
```
